```python
import pandas as pd
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder , MinMaxScaler
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

import numpy as np
```

```python
data_set=pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.train.csv")
```

```python
data_set.head()
```

```python
Y=pd.DataFrame(data_set["Churn"])
```

```python
columns=['Churn','gender','Partner','Dependents','PhoneService', 'MultipleLines',
       'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
       'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
       'PaperlessBilling', 'PaymentMethod']
```

```python
label_encoder=LabelEncoder()
```

```python
#transformo la varibaili categoriche in numeri
```

```python
for n in columns:
   data_set[n]=label_encoder.fit_transform(data_set[n])
```

```python
data_set
```

```python
Y=data_set['Churn']
```

```python
data_set=data_set.drop(['Churn','customerID'],axis=1)
```

```python
data_set.columns
```

```python
#transformo tutti gli elementi presenti nella colonna in interi
```

```python
data_set['TotalCharges']=pd.to_numeric(data_set['TotalCharges'],errors='coerce')
```

```python
#porto tutti i numeri presenti nella colonna totalcharges in un range tra 0-1
```

```python
transform=[
    ['min_max_scaler',MinMaxScaler(),['TotalCharges']]
]
```

```python
colum_transf_min_max=ColumnTransformer(transform,remainder='passthrough')
```

```python
X_data=pd.DataFrame(colum_transf_min_max.fit_transform(data_set))
```

```python
X_data.columns=['TotalCharges','gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod',
       'MonthlyCharges']
```

```python
#inserisco i valori mancanti nella coonna TotalCharges
```

```python


transform_imputer=[
    ['imputer',SimpleImputer(strategy="most_frequent"),['TotalCharges']]
]
```

```python
colum_transf=ColumnTransformer(transform_imputer,remainder='passthrough')
```

```python
X_data=pd.DataFrame(colum_transf.fit_transform(X_data))
```

```python
X_data.columns=['TotalCharges','gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod',
       'MonthlyCharges']
```

```python
X_data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_data,Y)
```

```python
rfc=RandomForestClassifier()
```

```python
 rfc.fit(X_train,Y_train.values.ravel())
```
